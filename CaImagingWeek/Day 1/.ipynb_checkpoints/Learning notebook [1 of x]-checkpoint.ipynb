Calcium imaging in neuroscience refers to a technique where you can actually *watch* neurons firing away. This is great, because humans are very visual animals (did you know over half of your brain is devoted to vision?!), so we like to be able to "see" what we're measuring.

+ fluorescent calcium sensors
+ 2-photon microscopy
