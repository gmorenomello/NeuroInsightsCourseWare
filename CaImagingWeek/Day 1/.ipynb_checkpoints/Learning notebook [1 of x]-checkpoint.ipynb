### Calcium sensors
What is the big deal about calcium anyway? 
<br> <br>
Imagine you wanted to visualize neural activity and kick off the field of optical physiology (ophys). How would you start? Well, you know that the unit of neural activity is an action potential. During an action potential, the main thing that changes is the voltage of the cell membrane (that's what we measure in electrophysiology!). So perhaps you decide to make a fluorescent voltage sensor. That's a great idea, and many people are working on exactly that, but so far it's been hard to find a reporter that has a good enough signal-to-noise (SNR). 
<br> <br>
Another approach could be to think about what molecules have a big change in concentration around the action potential. You might think first of sodium (Na+) and potassium (K+) since these are the big engines for the change in the membrane potential. However, the number of Na+ and K+ molecules that flow across the membrane during an AP is actually rather small and doesn't cause an overall change in concentration. But can you guess another ion that _does_ have a big change in concentration?



cha-ching! you got it. Calcium. Ca++. It turns out that, aside from building strong bones, calcium is crucial for neural signalling. It's involved in heaps of molecular cascades, from releasing neurotransmitters into the synapse to gene transcription to activity-dependent synaptic plasticity (aka learning!). And, most importantly for us, the concentration of Ca++ during an action potential is 10-100 fold [1]. That makes it a great candidate to serve as a marker for neural activity.

Now, all we need is a fluorescent calcium sensor. At this point, let's welcome the crystal jelly, or _Aequorea victoria_ to the story.

This little jellyfish found in the Pacific Ocean along the US coast was a gold mine for biomedical science. It is the origin of the ubiquitously used GFP. In addition though, it led to the first extraction of a different kind of fluorescent protein - Aequorin [2]. This protein is remarkable because it only fluoresces when it binds Ca++ ions.
