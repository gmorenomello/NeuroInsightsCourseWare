Programming is all about problem solving. It is the art to develop a set of step-by-step instructions that can transform a set of inputs into a desirable output, so we can solve a problem. The best analogy is developing recipes for cooking a specific dish. Defining a problem becomes harder when some assumptions are not shared. Because the computer knows nothing about the world, simple assumptions like "having to open a jar before scooping jam from it" are not known, and have to be defined. 

In fact, the fundamental knowledge of a computer is very limited. It only knows that it circuits that are either passing energy or not, like switches turned on or off. This is the basis of what you know as BITs, the fundamental unit of information that can be represented by 0 or 1, False or True.

Computer scientists in the past developed clever ways of combining sequences of may BITS to create dictionaries to represent more values. For instance, with one bit they could represent only two values, but with two bits they could represent 4 values, and with 3 bits 8 values. Essetially, for every new bit in the sequence you double the number of possible permutations, and therefore the number of possible entries in the dictionary.

This is essentially how computer scientists in the past defined how values are represented in computers. A given permutation of bits means a number or a character. For instance, a dictionary of all characters in romance language plus digits can be stored in a dictionary of 8 bits, or $2^8$, or 256 entries. This strategy was so widely used that 8bits gained its own name, Byte. And you may recognize that from metrics of memory in your computer.

It must be clear that this is not the only possible or used mapping between values that are comprehensible to humans and series of bits that are comprehensible for the computer. Computer scientists use many different of these "dictionaries" to describe their problems and consume as little memory as possible. 

For instance, it is possible to use a Byte to represent numbers from 0 to 255, instead of characters. Or the same byte can be uset do represent values between -128 and 127, the same 256 possible entries. In the same way, more bits can be used to represent more different characters, for instance the chinese characters, or a wider range of numbers.


### ABSTRACTION
This lead us to a second important concept in programming, Abstractions. Abstraction means to get low-level detailed description, and simplifying them with easy to use representations. 

Converting sequence of bits into letters is a simple form of abstraction, as it allow us to write a text without having to think which sequence of bits represents the letter G, for instance. 

One Byte offcourse is not the limit in this regard. There is an other form of mapping called Unicode (UTF-8) where up to 4 bytes are used to represent characters every character in every language know by humanity, plus the Emoticons.

Offcourse we are not limited to represent characters. We can also represent colors in the context of an images. Images are sequences of bits. Pixels. That can be organized in an bidimensional array (matrix) for each color. One for Red one for Green and one for Blue. A Byte in each matrix represents the amount of color that one particular pixel has. By combining the three, they become the color of the pixel. For example: RGB [73, 88, 255]. 

As for the text, there are other mappings for color, like: HSV which stands for Hue, Saturation and Value.

Going further, we can combine images into videos, as videos are just sequences of images, that for some reason we call frames. Hence: Videos are abstractions of images; Images are abstractions to pixels; Pixels are abstractions of bits.

By using videos as abstractions, you don't need to work with the bits, but with scenes and moments. That are easier for you to understand and use.

In the same way that a map represents the path and the geography of a land in a simpler way. Highlighting some aspects and ignoring others, making it easy to find your path in the land. Abstractions of the data, allows you to map your problem in a simpler way, so it becomes easier to solve it.

### Algorithms
To solve a problem we usually start from a set of inputs or data, perform some steps and operations to transform this data and generate an output. In the context of programming the output might be a new number, a sentence on the screen, a graph or something that we want the computer to do.

In the same way that we defined the data for the computer, we also need to define a series of step-by-step instructions that the computer can follow to solve the problem. This is what we call an algorithm, a recipe to solve a problem. 

Algorithms do not depend on a computer to be performed, and they can vary in performance (speed) and correctness (if it solves the problem or not). Which makes computer scientists invest a lot of time and brain power to develop better algorithms that can get to the right result, faster, consuming less energy and memory.

To exemplify, let's examine how you would find out an word in a dictionary. We could go page by page, from the first to the last one, looking for the word we want. That would certainly reach the correct goal. But it would take a lot of time, especially for a word starting with the letter Z. 

We could improve the speed starting from the nearest end of the dictionary and by going every 10 pages, but in this case we could miss the page with the word. 

The missing page problem could be corrected by going back if we noticed that we passed over the page with the desired word.

Well, lets see an alternative that is much faster. We can go to the half of the dictionary and check if the name is in the part before or after. We chose the part that the word is likely to be, and repeat the first step, over and over until we find the desired word. This algorithm is much simpler to describe, it is guaranteed to reach the desired goal and does it in less time than the algorithm before. 

In computer science we say that this last algorithm is less complex than the first one. Your goal when developing your programs is to build series of step-by-step instructions, that are as simple as possible and efficient as possible. If you are curious, you can look for Algorithm Complexity and Big O to find out how the effiency of algorithms can be measured.

### Programming definition
So this is what we do when we program. We create Algorithms with minimum complexity using abstractions to transform inputs into outputs useful to solve a problem. 

To solve a problem, we only need just a few abstractions that describe the operations that need to be done, and list them in a sequence. In the video you saw, there were some few operations that are very common in programming like.

**Variables** - names that describe some value to be kept in the memory.
**Boolean expression** - questions and checks that test if something is true or false.
**Conditions** - bifurcations in the step-by-step that allow to do something different depending on the situation, the values of the variables or boolean expressions. For instance, "if this then that, else do otherwise"
**Loops** - sequences of steps that have to be repeated many times until a given condition is achieved.
**Function** - names that encapsulate and abstract a series of instructions that do something. Like a mini algorithm.
