### 3.1 Jupyter notebook introduction
The **Jupyter notebook** is a very interresting IDE because it allows you to organize your code in notebooks. To start the jupyter notebook you have to first open the **command line** or **terminal** in your operating system. 

To get to the command line on Linux or MacOS. You just need to find the app called **Terminal** and open it.

On Windows, you have to open the **command line** by typing **cmd** in the search engine and double clicking the first result of the search.

To start the jupyter notebook you just need to type: 

```python
    jupyter notebook
```
on MacOS or linux, and ... 

```python
    jupyter notebook
```
on Windows.

The **jupyter notebook** will open your web browser because it runs entirelly as a web page. You can use Jupyter notebook to run code in R or Octave (the open source twin of Matlab), but that is outside of the scope of this course. Please take some time to familiarize with the interface. There are 3 main Tabs on top, File, Running and Clusters. File is for the files inside the current folder, Running is for the files that are running right now, and Clusters is to visualize the available custers of computers. Clusters are groups of computer that join their processing power to run software faster, ignore this last one for now.  Try to navigate to the folder where you are going to put your code.

To start programming you have first to chose a folder where to store your scritps. Then you can create a **new notebook**. You must chose the kind of programming language you must use. Chose **Python 3**.

The **notebook** is dividend in cells of contents. In each cell you can either insert a code that you can execute, or a formated text, which you can use to document your script. This is especially useful in data analysis, and because of that, widely used by companies.

Remember to rename the notebook as soon as you open it. In this way your scripts remain organized. 

To enter and edit a cell, you must double click it or press **Enter**. Here you can type the code or comments however you like. Type:
```python
    print(13)
``` 
At this point you can press **Shift + Enter**. This will run the code and change the selection to the cell below. Notice that it displays the number 13 in the cell below.

If you want to exit the cell, and select the cell. Press **Esc**.

The [* ] on the left corner means you are running the code. If it changes to a Number it means that if finished running the code in the cell and the number reflects the order that you executed the cell in the notebook. 

While the cell is selected and out of the editing mode. You can press **A** to add a cell above and **B** to adds a cell below.

By typing **D twice quickly** you can delete the cell.

If you need to figure out what certain functions do. Type ?name_of_function and the function help description will show up on screen. To see how it works, type:
```python
    ?print
```

Alternatively, you can type the mame of the function and press **shift+Tab**.

If you need to access the **terminal / command line** from within the notebok, Press **!command**. This will allow to execute the command as if you were in the terminal or the prompt.

A very helpful command is to press **commandKey + F** or **ctrl + F** . It looks for all the instances of the word you are looking for in the cell

A very special feature of Jupyter notebook is that it allows you to **anotate** your code and output very well organized reports of your code. By pressing the letter **M** when a cell is selected, you can change some of these cells into well formated texts, using **Markdowns**. Markdowns are especial sybols that allows you to format your text in the same way you do with a document in a **text editor** such as **Word** or **Pages**

### 3.3 Spyder
Sypder is a graphical IDE for python shaped to imitate MATLAB. It has all of its main components: 
- An Editor
- A command line console
- A help (Ctrl/Command + i)
- A File explorer
- Outliner
- Integrated Debugger

You can start a new project by selecting *Projects-> New Project...*. 

Similar to MATLAB, you can separate the code into cells by lines starting with either:
- `#%%` (standard cell separator)
- `# %%` (standard cell separator, when file has been edited with Eclipse)
- `# <codecell>` (IPython notebook cell separator)
- `#### YourComment`to add a comment to the outline


Learn from https://docs.spyder-ide.org/index.html

Spyder has integration with **pdb** a python debugger. Read more at https://pythonconquerstheuniverse.wordpress.com/2009/09/10/debugging-in-python/.
