
# Oh no!

Something is not working how you think it should or you cant quite wrap your head around something. Don't call the fire brigade quite yet, there is a few things we can try first. 

* Read the Documentation 
* Checking warnings
* Googling the issue
* Stackoverflow
* Inspecting the code or using the debugger
* Update your dependencies
* GitHub issues
* Looking at the Forum
* Consult the oracle (Rubberduck) (not a joke, it really can work)


![Oracle](../Resources/memes/duck.jpg)


## Documentation

If you are just trying to test a framework, having and issue with one, or trying to figure out what is possible, then these are all good reasons check the documentation. Most programming languages and framework has documentation although of varying quality, the only thing worse then a poor documentation is no documentation. 

If your trying to achieve something specific see if they have an example

If you are having a issue, try to follow the recommended setup and potential examples to verify your setup. Documentation also often document parameters that can be helpful to identify when you've made a wrong assumption.

## Warning!

These yellow labels you see everywhere that don't actually stop you from progressing? They might actually be saying something important in your context and even give you tips for improvement. 

## Tips for googling: 

Well duh, of course you've googled you don't need me to https://letmegooglethat.com/?q=How+to+exit+Vim for you. 

Yet, there is more to googling the just typing some keywords and rolling the dice. 

Not finding something you think should exist?

Depending on the problem you can try:

Rephrasing your query, perhaps change your keywords into a question or vice versa.
Use some synonyms
Searching for a error? make sure to generalize your query by removing things like classes in the error stack or message
Advanced search (https://www.google.com/advanced_search) or search operators (https://www.webfx.com/blog/google-2/google-advanced-search-operators-cheat-sheet/), quotation marks can be especially useful. 
Try another method, not everything can be found using google.

## Stackoverflow.com and the stackexchange

Welcome to the 8th wonder of the internet stackexchange and stackoverflow. Stackexchange is the larger collection of subsites that work as a database of Q/A about specific topics. The largest and perhaps the most relevant one is stackoverflow, but do check out the full list:

https://stackexchange.com/sites perhaps you'll find something interesting, like https://psychology.stackexchange.com/ 

You might often land here if you are goggling coding problems but it could also be worth to seek the website out directly, because again not everything can be found using google and sometimes you have to be the first to ask the question.

Tips: 

The integrated search bar features tag searching, so if your interested in a question about matplotlib, it might be helpful to tag your question [matplotlib]
If you think there should be a Q/A for your issue but cant find one? try searching for just a tag and ordering by frequency or votes. 
Ask a question, but before you do, understand what the guidelines are: https://stackoverflow.com/help/how-to-ask and steel yourself to the possibility that stackoverflow can be harsh or mean or even outright toxic, though there is some information that might help you understand why https://meta.stackexchange.com/a/342788

## The Debugger

A debugging tool is usually integrated in most IDE's and learning to harness its power is well worth the effort. Print("here") might get you far, but the debugger can get you further!

Learn to use the debugging tool of your preferred IDE, set breakpoints and step through your code while looking at the state of the variables. 

## Update/revert your dependencies 

Software is often changing, and many projects are maintained actively, by the time you discover an issue it might already be solved so an update can be very useful, but the opposite can also be true, perhaps a revert is in order if libraries do not want to work together. 

## GitHub issues

Having a problem where library X does not work as it supposed to? Its entirely possible that there is a issue with library X. Very few projects have no issues, if they do, they probably just haven't been discovered yet. It can be well worth it to check if there is a open issue with library X, perhaps you might even find a hack to solve the issue, or information on which version of the library you can revert to in order to avoid it. 

## Forums

No, its not 2001 again, but a decent amount of frameworks have dedicated forums, it might be worth it to check for answers there, or even ask a question.
