## What is data science?

**Data Science** can be thought as a field of study that pursuits to *understand and to explain the structure of data*, to make *predictions* from this understanding using any means necessary, mostly by *using algorithms and principles from computer science, statistics and machine learning*. In other words, it aims at transforming observations into insights, predictions and understanding.

Every science uses data to get insight about the phenomenon it wants to understand. So, what is special about Data Science? The difference, between data science and other sciences, is that in data science, the data themselves take the center stage and become the subject of study. This means that within an explicitly data-centric framework, the source of the data does not matter much. What matters is the kind or type of the data. ## From observations to measures to metrics to statistics

To extract insight from nature we need to make sense of our **observations**. In science, we observe different phenomenas and we can describe them in many different ways, including long text reports. 

But, ideally we want to describe observations in a way that is economical, as objective as possible and in a way that allows us to compare multiple observations. So we can conclude something from them. 

This is the advantage of Quantitative methods. We attribute values to magnitudes of our observations and use them as **measures** of our observation. Generally in neurosciences we do not have direct access to what information the brain is processing, so we use measures of observable variables as **metrics** (standardized quatifiable representations) of underlying, less observable phenomena.

Two classical examples are Attention Load and Learning. Just by looking at someone it is hard to tell if they are paying attention or learning something. To do so, we have to put the participant or subject under a situation where it becomes possible to measure through their behavior the attention or learning.



In the case of Neurosciences, we want to infer something about how the brain works and implements behaviors by measuring observable phenomenas. 

Along the course, we are going to explore multiple ways of encoding observation into quantities and doing several operations on them to describe, sumarize, and compare them, as well as generate predictions.

#### Missing Data
Missing data or missing values occur when no data is stored for a certain oservation in a variable.
Missing data are common occurrence in most datasets
Missing data can have significant effect on the conclusions that can be drawm from the data

**Missing data causes** A value is missing because it was forgotten, lost or not stored properly. If the value of the variable do not exist (for example dividing by 0), not found or not identified.

Many libraries do not support missing data so a decision must be made if the observation will be ignored or if a data imputation technique will be employed.
Missing data imputation affect your models and the insights that you can take from them So you will have to have that into consideration.

**Missing data Mechanisms** understanding the missing data mechanisms may help us to choos the right missing data imputation technique. 
- **MCAR** - Missing data completely at random: when the probability of being missing is the same for all the observations. There is no relationship between the data missing and any other values observed or missing within the dataset. Thus, disregardign those cases would not bias the inferences made. 
- **MAR** - Missing data at random: The probability of missing an observation depends on available information. For instance, if it is more likely to have missing data from female participants than male ones. This might be a good idea to include the independent variable as a control to balance the missing points.
- **MNAR** - Missing data not at random: There is a mechanism or a reason why missing values are introduced in the dataset. For instance, if people do not reply to the depression survey because they are depressed. In this situation it could be useful to flag these observations, because they allow to predict the outcome. The missing values are predictive of something.

## Cardinality
The cardinality of a set $ S $ is the number of distinct(unique) elements in the set. Which is notation is $ |S|$. Thus if $ A = \{1,2,3,2\} $, the cardinality of A will be $|A| = 3 $. The same applies for categorical sets where $G = \{male, female, male, male, female, female, female\}$ the cardinality will be  $|G| = 2 $.

The values of a categorical variable are selected from a group of categories or labels
The number of different labels/categories is known as cardinality.
**Cardinality effects** many libraries and models do not operate on strings. This means that it is necessary to convert categories into numbers to operate them in mathematical models. Categorical encoding techniques must be applied. One example of such techiniques is one-hot encoding. and it is used in several methods to build something called a design matrix.

High cardinality might also lead to uneven distribution between train and test sets, because some categories might fall only in one of the sets. Leading to over-fitting in tree based algorithms and operational problems.
Variables with too many labels tend to dominate over those with fewer labels, particularly in tree based algorightms.
A big number of lables within a variable may introduce noise.
Reducing cardinality may help to improve model performance.
- strings(categories) need to be encoded as numbers to be used in models
- high cardinality may cause several over-fitting and operationalisation problems
- Reducing cardinality may hel to improve model performance.

**rare labels** are more likely to be found in high cardinality sets. They cause the same effects of high cardinality. As they introduce noise in the data. One must consider if rare labels should be combined in a meaningful super category or discarted from the dataset.

### Linear Model Assumptions
Linear models make the following assumptions about the independent variable (X)
- **Linear relationship** between the variables and the target: $ Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + ... + \beta_nX_n$. Where $\beta_0$ is the bias or intercept, and the other betas represent the slope. And the other Linear relationship can be assessed with scatter plots. Sometimes non-linear transformations of the variables and the target improve the linear relationship.
- Multivariate normality: This happens where every independent variable follows the Gaussian distribution. Normality can be assessed with **histograms** and **Q-Q plots**. Normality can be statistically tested, for example with **Kolmogorov-Smirnov** test. When the variable is not normally distributed a non-linear transformation (e.g., logarithmic transformation) may fix the issue. 
- No or little co-linearity: Multicolinearity occurs when the independent variables are correlated with each other. Multicollinearity can be assessed with a correlation matrix or the variance inflation factor (VIF).

### File name conventions
The first stem in data organization id deciding on a file name convention. This helps on communication, search and maintenance of the code and database. Especially when more than one person is working in the project.

A good convention should have in mind scripts and applications so that automation is easier to implement. For instance, if the naming has the days of the weak, it is better to just use the three first letters than the whole day name, which varies in size.

#### Date and time in a file name
Useful as context and organization tool. Here are the benefits to use Date and Time as filenames.
- Date is useful information. Remember that file management systems order alphabetically and numerically. So the format YYYY_MM_DD (Year, Month, Day) will be sorted in a more meaningful way than DD_MM_YYYY or MM_DD_YYYY.
- File names are almost guaranteed to be unique. If you want to guarantee uniqueness, you just add time HH_mm_ss hour minutes seconds to the filename.
- Date and time information will be retained when the file is copies, moved or backed up.
- It is easy to automate and write scripts that include date infomration.
- Date convention is easy to implement accross different systems and languages.

#### Useful filename titles
Filenames should be useful, short and descriptive. Good tips are:
- Uniquely identify the data source system. This allows to identify and quickly deal with data that is corrupted by falty equipment.
- Uniquely identify the instance of the data (Project, Experiment, Subject, etc). Macro groups first, subgroups latter.
- If the data is not raw, identify which script and version was used to generate it from the raw.
- Avoid describing the data type. Names such HighPassVoltageRecordings make it hard for you to add other variables into the file. And if you do it will be misleading.
- Prefer names that are easy to parse. Plexon789_TMaze_Rat4 is easier to parse and read than plexon789tmazerat4
- Preferentially, fix the number of letters for each information component in the title. Example Rat4, Bir3, Fis2, Mou5, respectivley for Rat, Bird, Fish, Mouse.
- Have in mind the multiple data sources that you will have to consolidate. If you have eye tracking and EEG systems you might prefer lead the title of the file by Experiment_Subject_System instead of System_experiment_subject. In this way all the files belonging to the same experiment and subject will be close to eachother.
