```python
# Broadcasting
#You can create a new column on the fly by calling a new column name and adding a value to it
# the value is broadcasted to the entire column without needing to use loops.
import numpy as np
df1['Fee'] = np.nan
df1

#You can also broadcast other values other than numbers
height =[189, 179,190, 198, 176, 175, 181]
data2 ={'height': height, 'sex':'M'}
df2   = pd.DataFrame(data2) 

# CHANGING LABELS
# we can change the columns label of a data frame by changing the .columns attribute
df2.columns = ['Height(m)','Gender']
# we can change the index labels of a data frame by changing the .index attribute
df2.index = ['A','B','C','D','E','F','G']
# The number of elements in the columns and in the index must be precise
```
