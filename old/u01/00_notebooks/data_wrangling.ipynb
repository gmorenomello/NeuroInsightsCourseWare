# Pandas
Pandas makes use of two different data structures
- Series - 1D collection of values with labels and indexes (starting on 0)
- Data Frames - 2D collection of Series where each collumn is a Series

## Series
Represents data in 1D like a vector

```python
import pandas as pd
# Convert dictionaries to Series
my_dictio = {'a': 3, 'b':56, 'd':999 }
pd.Series(my_dictio)
```

```python
# type ?pd.Series to see how the class pd.Series works. 
# ? can be used in the front of any function or class to show its documentation
```

```python
my_values =[12,"fifo", 36.5]
my_indices = ["age","name", "temperature"]
my_serie = pd.Series(my_values,my_indices)
my_serie
```

```python
# returns a series with only the values and indices in a particular index label
o = my_serie.loc[["age", "temperature"]] 
o
```

```python
# returns a series with only the values and indices of a particular value, 
# good for changing the order of values in a Series.
a = my_serie[[1,0,2]]
a
```

```python
my_serie.iloc[1:] # gets the value of the specified index.
```

```python
print( "age" in my_serie)
print( "Age" in my_serie)
print( "fifo" in my_serie)
print( "fifo" in my_serie.values)
```

```python
# Building Dataframe
import pandas as pd

# You can crea data frames using csv files
fn = r'/Users/gmorenomello/Documents/Python/181101_STEFANO/Composite Circle6.txt'
rois = pd.read_csv(fn, index_col=0,sep=' ')
#print(rois)

# You can also create data frames using dictionaries
data0 = {'weekday':['Sun','Mon','Tue','Wed','Thu','Fri','Sat'],
        'city':['Paris','Oslo','Rio de Janeiro','Lisbon','Poznan','London','Madrid'],
       'visitors':[12,2,33,43,12,32,93]}

df0 = pd.DataFrame(data0)
# The keys of the dictionary are used as column headers (labels)
# by default the rows are indexed from zero

# You can also build DataFrames from lists
# What you need to do is to combine the lists into a dictionary
weekday = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
city = ['Paris','Oslo','Rio de Janeiro','Lisbon','Poznan','London','Madrid']
visitors = [12,2,33,43,12,32,93]
# For that you need to define two additional lists
list_labels =['City','Visitors','Week day'] # for the column label header
list_cols = [city,visitors,weekday] # for the order that the lists will be added to the columns
# Now you combine both lists into a list of tuples
zipped_list = list(zip(list_labels,list_cols))
# Finally you transform the list of touples into a dictionary
data1 = dict(zipped_list)
df1 = pd.DataFrame(data1)

```

```python
# Broadcasting
#You can create a new column on the fly by calling a new column name and adding a value to it
# the value is broadcasted to the entire column without needing to use loops.
import numpy as np
df1['Fee'] = np.nan
df1

#You can also broadcast other values other than numbers
height =[189, 179,190, 198, 176, 175, 181]
data2 ={'height': height, 'sex':'M'}
df2   = pd.DataFrame(data2) 

# CHANGING LABELS
# we can change the columns label of a data frame by changing the .columns attribute
df2.columns = ['Height(m)','Gender']
# we can change the index labels of a data frame by changing the .index attribute
df2.index = ['A','B','C','D','E','F','G']
# The number of elements in the columns and in the index must be precise
```
