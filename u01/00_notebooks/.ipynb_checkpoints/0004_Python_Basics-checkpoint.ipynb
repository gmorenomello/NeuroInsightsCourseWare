### 4.2 Arithmetic Operators

Arithmetics operators:
* \+ addition
* \- subtraction
* \* multiplication
* \/ division
* \*\* power
* \% remainder
* \/\/ round down division
Order of operations apply and parenthesis can be used too

The arithmetical operators will have different effects depending on the data type of the value.

````python
print(12/3)
print(-7.1/3)
print("Ingrid" + " Ferdinando")
print("Ingrid "*9)
print(True * False)
````

In Python,this is called an **expression**, which is the most basic kind of programming instruction in the language. That means you can use expressions anywhere in Python code that you could also use a value.

You attribute a value to a variable by using the = (assignment) operator.
```python

```

> **please note that in math assignment is annotated as := and not =. Assignment means that the value on the right will replace the value that was previously being represented by the variable**

You can assing many values to many variable names simultaneously.
```python 
   name, age, gender = 'Jane', 25, 'Female' 
```


**Variable naming good practices:** Python has some formalities on how to name variables. 
1. They should describe what they represent.
2. They cannot start with a number.
3. They should be in lower case.
4. You should use underscore to connect words.

For now your programs are short. But as you progress your softwares will become big and complex. When that happens your variables might be instantaneously used in the next lines of code, or used by distant parts of your code. The wider the scope of use of this variable, the longer and more descriptive should be its name.

### Flow Control
#### Flow charts
**Flow control statements** are parts of the code that can evaluate the state of some value or expression and decide which instructions to execute under which conditions.

These flow control statements directly correspond to the symbols in a flowchart, so I’ll provide flowchart versions of the code discussed in this lesson. The flowchart immitates a Bouncer in a nightclub. The flow chart describes the decision path made by the arrows from Start to End.

#### Boolean values

The **Boolean data type** has only two values: **True** and **False**. (Boolean data type is named after mathematician **George Boole**.) The Boolean values True and False **lack the quotes** you place around strings, and they **always start with a capital first letter**, with the rest of the word in lowercase.

Operators | Meaning

#### Boolean operators
The three Boolean operators (**and**, **or**, and **not**) are used to compare Boolean values. Like comparison operators, they evaluate these expressions down to a Boolean value.

- and means that both values must be True to return true.
- or means that at least one of the Boolean values must be True to return true.
- not flips the value. So if it is True it will be evaluated as False, and vice-versa.

#### Flow control statements
**Flow control statements** often start with a **keyword (if,elif, else)** followed by a part called the **condition** and a **collon** (:). All these parts are followed by a **block of code** called the **clause**. They usually take this format:

```python
age = input() 
if int(age) >= 23: # if is the Flow Control Statement
    print('can pass') # block of code 1
elif int(age) >= 18: # condition 2. elif stands for "else if"
    print('can pass if you have your parents with you') # block of code 2
else: # do this otherwise
    print('go to disneyland!') # block of code 3
```
**Conditions** are just how Booleans are called in the context of Flow control.
 
**Blocks of code** are lines of code grouped together by the same identation (spaces from the left margin). There are three rules for building a block of code:

1. Blocks begin when the indentation increases (usually tab or 4 spaces).
2. Blocks can contain other blocks.
3. Blocks end either when the indentation is decreased to zero, or a new block inside the current one starts.

Here is an other example:
```python
age = input() 
if int(age) >= 23: # if is the Flow Control Statement
    print('can pass') # block of code 1
else: # do this otherwise
    years_to_go = 23 - int(age)# block of code 2
    print('Come back in ',str(years_to_go), " years." ) 
```
The most common type of flow control statement is the if statement. An if statement’s clause (that is, the block following the if statement) will execute if the statement’s condition is True. The clause is skipped if the condition is False.
- The **if** keyword
- A condition (that is, an expression that evaluates to True or False)
- A colon
- Starting on the next line, an indented block of code (called the if clause)

The else clause is executed only when the if statement’s condition is False. 
- The **else** keyword
- A colon
- Starting on the next line, an indented block of code (called the else clause)

The else clause is executed only when the if statement’s condition is False and the condition following elif is evaluated True. 
- The **elif** keyword
- A condition (that is, an expression that evaluates to True or False)
- A colon
- Starting on the next line, an indented block of code (called the elif clause)

### Concurrent Future




````python
from concurrent import futures
def factorize(n):
    """Return the prime factors of *n*
    
    This is a **very** bad factoring algorighm.
    
    """
    
    n = int(n)
    if n == 1:
        return 1, []
    found = []
    
    for i in range (2, (n//2) +1):
        if n % i == 0:
            if all (i % j!=0 for j in found):
                found.append(i)
                
    if not found:
        return n, [n]

if __name__=='__main__':
    with futures.ProcessPoolExecutor() as pool:
        for number, factors in pool.map(factorize,range(1,10001)):
            print('{}: {}'.format(number, factors))
````

the pool.map method returns an generator (iterable object with all the results of all the processes started).
Also, the first argument of pool.map is the function. The function must be pickleable (not a function inside a function). And if you send a method, it will send a copy of the object (not a reference) to the process.


````python
from concurrent import futures
def foo(x, y):
    return y*x

with futures.ProcessPoolExecutor() as pool:
    for x in pool.map(foo, [1,2,3,4,5,6,7,8,9,0], [11,22,33,44,55,66,77,88,99,100]):
        print(x)
    
````

### Submit method

Instead of waiting for all the processes to be complete in order to collect the results. The submit method allows you to get information from the process as they are finished.

It is important to notice that you can also submit different functions to different processes. Each submit call is one call to an individual process. 

The submit method returns a *future* (IO) object that allow us to monitor and interact with the function that is being executed in the particular process.

````python
from concurrent import futures
def factorize(n):
    """Return the prime factors of *n*
    
    This is a **very** bad factoring algorighm.
    
    """
    
    n = int(n)
    if n < 2:
        return 1, []
    found = []
    
    for i in range (2, (n//2) +1):
        if n % i == 0:
            if all (i % j!=0 for j in found):
                found.append(i)
                
    if not found:
        return n, [n]
    return n, found

with futures.ProcessPoolExecutor() as pool:
    fs = [pool.submit(factorize,x) for x in range(1,10001)]
    for f in futures.as_completed(fs):
        try:
            number, factors = f.result()
        except ValueError:
            continue
        print('{}: {}'.format(number, factors))
````

````python
from concurrent import futures
class FooError(Exception):
    pass

def foo(x,y):
    if x>y:
        raise FooError
    return y-x

with futures.ProcessPoolExecutor() as pool:
    future = pool.submit(foo,4,7)
    future1 = pool.submit(foo,14,70)
    x = 0;
    while not future.done():
        # .done method returns true if the job is done
        print(x)
        x = x+1
    while True:
        try:
            data = future.result(timeout = 0.5) #look at "ascompleted"
        except futures.TimeoutError:
            print('Working....')
            continue
        
    try:
        data1 = future.result()
        data2 = future1.result()
    except FooError:
        data = None
print(data1)
print(data2)
````

to process multiple futures

````python
from concurrent import futures
class FooError(Exception):
    pass

def foo(x,y):
    if x>y:
        raise FooError
    return y-x

with futures.ProcessPoolExecutor() as pool:
    running =[pool.submit(foo,5,8),
              pool.submit(foo,6,9)]
    
    while running:
        done, running = futures.wait(running, timeout=2.5)
        for future in done:
            data = future.result()
            #do whatever you want to do with the data.
````

### Looking for inconsistencies in the data
Always look in the last column. Inconsistent data in the last column is usually a symptom of problems in other columns like: missing data, poorly formated data, missing field for data, error in splitting data when importning.

Look for empty columns.
If a column seems empty, check if is there any non-empty value on it and whether it makes sense. Also check if the second to last column is violated. If the column is completely empty, you might decide to remove it completely.

Are the column titles named consistently? Same Caps format, same number of letters, etc. Is there a missing title? Is any title in the wrong place? You can see this by looking if the title describes well the data that is in the column.

Check for the data-type of the columns. Try to convert to the correct data-type. Try to identify values that cannot be converted into the desired data-type. These values might reflect errors in the importing of the data or during the acquisition.

In every column, look for values that violate the format that the data should be in. See if you can fix it.

#### Missing data 
Look in the variables that are most critical for your analysis. Is there any missing value? There are few ways of handling with missing data:
* if categorical: you can eliminate the data point or create an undefined or unknown category.
* if numerical: you can eliminate if you have too many data points or you can replace it for a central tendency measurement. Like the mean, the median or the mode. The one that you chose depends on the analysis you want to make and on the distribution of data over all.
* if time-series: You can eliminate, but usually in time series the steps in the future depend on the steps in the past, and eliminating causes more trouble than in other types of data. So a better strategy is to use a running mean/median and interpolate the points.

To eliminate a data point you can either not include it into your final table or you can replace the value with NaN = Not a number. Value.

#### Outliers
Outliers are values that drastically deviate from the average. Outliers are tricky because they can be true outliers (exceptional individuals in your sample that actually deviate from the population) or they can be error derived outliers. The only way to tell one from the other is to look at your data and as yourself "does this value make sense for this observation?" . Sometimes the answer is obvious, for instance, one individual may have an anual income of an entire village, like a millionaire from Norway that decided to live in a village in Thailand. Sometimes the only way to know is by knowing how the data was acquired, or talking to the specialist that acquired it. 

If the data point does not make sense, it is probably a useless outlier. In this case, you have to eliminate it and deal with it in the same way you would with the missing data.

### Vector Norms
Ways of measure the magnitude of vectors. This is important to understand how some machine learning algorithms operate and what they give you as an output.

There are two Vector norms: L2 - Norm and L1 Norm. 
Given a vector with two components: x = (a,b). 

L2 - Norm: is usually noted as !!xArrow!!2 = sqrt(a^2+b^2) in other words. it is the cartesian distance.

L1 - Norm: !!xArrow!!1 = !a!+!b!

Now imagine that you want to plot all the possible points on a cartesian plane where the L1 or L2 norm is equal to 2.

The L2 Norm will describe a circle. 

The L1 Norm will describe a square. Because to keep the equality, as one number increase the other should decrease in equal proportion.

To generalize for multiple dimensions. All you have to do is to add more components. L2 = Sqrt(a^2+b^2+c^2+...+z^2) L1 = !a!+...+!z!



## Statistics





### R-Squared: Coefficient of Determination
R2 or R Squared is a measure that expresses the amount of explained variation of a model.

**SST - Sum of Squared Total:** it is the sum of the squared deviations from the mean (named as letter bar). In other words. You take the mean, you subtract each point from the mean. You potentiate each difference to the second power. Than you summ it. This describes the total amount of variation.

**SSE - Sum of Squared Explained variation:** This describes how much your prediction function (named as letter hat, example ê) deviates from the mean. Or how much of the variation is explained by your prediction. In other words. You get each point of your prediction function, subtracts it from the mean, elevates it to the second power (square it) and than you summ it.

**SSR - Sum of Squared Residual** IMPORTANT, be careful, sometimes it is called SSE (sum of squared errors). This describes the amount of variation that is not explained by your model. To calculate it you subtract each datapoint (response) from your prediction function. you square each one of the differences and than you sum all the results together.

Hence the total variation (TV) is equal to the *variation of the model* plus the *variation of the error*.

**R2 = 1 - SSR/SST**

SSR/SST is the total fraction of variation that is not explained by the model. By subtracting it from 1, we are getting the complementary probability. Hence the complementary probability is the proportion of explained variation.

If the model is linear SST is equal to SSR + SSE.

### Linear Regression 

Simple regression model is Response **Y**(Dependent variable) equals the **BetaNot**(the intercept) plus **Betan**(a coefficient) times **Xn**(Independent Variable or predictor) plus the E (error term). Where each Xn is one independent variable and each Bn is one coefficient for that variable. Error is a buffer that allows the prediction to exist.

A simple linear regression will generate a coefficient table. Where every variable (feature or predictor) has its own standard error, coefficient, t-value and P-value. The P-value indicates which variables are significant at a level of significance (0.05 for example) and if it is, that means that it can be included in the model.

The interpretation for the coefficient means that for every change in the variable unit, you change the response function by the coefficient amount, given that the other independent variables are held constant.

#### Ordinary Least Squares Regression
OLS is a method for estimating the unknown paramenters in the linear regression.

The goal of the OLS is to minimize the differences between the observed responses in some arbitrary data set and the responsed predicted by the linear approximation of the data.

You do that by measuring the distance between the points in the arbitrary dataset from the function you want to fit. You square these distances. The squaring has two effects, 1- it makes all the differences positive, and 2- it penalizes more the biggest differences as those will count more. Finally you take the average of all these squared distances and this is your measure of error. OLS tryies to minimize this value.

One way of visualize it is by seeing squares with sides equals all the difference and the MSE (mean squared error) is the average area of those squares. The smaller the squares. the better the model.

**Gauss markov theory** states that the OLS regression is the (BLUE) - Best Linear Unbiased Estimator. is the smallest variance possible.

#### ERROR
The sum of squares are a representation of the error in our OLS regression model.

When we discuss linear regression models, prediction errors can be decomposed into two main subcomponents we care about: Error due to **"bias"** and error due to **"variance"**.

##### Bias and variance trade off
**Error due to bias:** The error due to bias is taken as the difference between the expected (or average) prediction of our model and the correct value which we are trying to predict.

**Error due to variance:** The error due to variance is taken as the variability of a model prediction for a given data point. Imagine that you can repeat the entire model building process multiple times. The variance is how much the predictions for a given point vary between different realizations of the model.

Use the metaphor of the target. low variance + low biase = all shots in the target, high variance + low bias = shots distributed around the center, high bias + low variance = shots grouped outside of the bullseye, high bias + high variance = shots all over the place

There is a tradeoff between a model's ability to minimize bias and variance. Understanding these two types of error can help us diagnose model results and avoid the mistake of over- or under-fitting.

The sweet spot for any model is the level of complexity at which the increase in bias is equivalent to the reduction in variance.

GraPH( Y= error , X= Model complexity, Bias2 decreases exponentially with growth in complexity (number of parameters), variance grows exponentially with grow in complexity. Total error is a U curve. The point where the two curves meet at lowest is the optimal spot.

### Regularized Regressions
Designed for higher dimension models. 
It helps to overcome the overfitting issue of the model.
It helps to keep the parameters **regular and normal**
Generally we do not want models with parameters with huge weights. Because small changes in the data will cause a large change in the prediction. This reduce the chance that the model will generalize.

What regularization does is to put a boundary on how much the weights can grow relative to each other and tries to reduce the contribution of the features that are not relevant close to zero.

In the end the model should have less features differend from zero and even the ones different from zero should have low coefficients.

How does regularization does this?
Min((Y - (Bi * X) ^ 2) + (lamda / 2) * sum (!Bi!) ^ p)
**Min((Y - (Bi * X) ^ 2)** OLS
**(lamda / 2) * sum (!Bi}!) ^ p)** Regularization term
- Bi = weights
- power of p = 1,2 etc
- lamda = penalization term.

It is still a convex optimization problem. In other words, you are not only minimizing the error, but also the sum of the weigths. 

Because some of features will be zero, the regularization help us to select some features. Because the model with a feature with zero weight is as good as a model without that feature at all.


Is it possible to find a model with even lower bias and variance than the OLS?

There are two types of regularization L1 and L2. 
It is important to understand loss function. The loss function is the sum of the difference between our prediction and the actual data.
L1 is the module of the loss function. or the Least absolute regression. LAR
L2 is the squared of the loss function.

L1 = Lasso regularization: 
Min((t(Xj) - sum(wi * hi(Xj)) ^ 2) + lamda * sum (!Wi!))

L2 = Ridge regularization: 
Min((t(Xj) - sum(wi * hi(Xj)) ^ 2) + lamda * sum (Wi^2))

How to find the optimal weights for L1 and L2?
Prediction accuracy tends to be better with L1 or L2.

#### Shrinkage estimators.
Lets consider something that initially seems crazy. We will replace our OLS estimates Bn with something sligthly smaller.
bprimen= (1 /(1+lambda)) betan

if lambta is zero, we get the OLS estimate back; if lambda gets really large, the parameter estimate approaches a minimal value of zero.

lambda is refered to as the **shrinkage estimator** (ridge constant) or the penalizing term.

In principle, with the right choice of lambda we can get an estimator with a better MSE

The estimate is not unbiased but what we pay for in bias, we make up for in variance.
https://www.youtube.com/watch?v=ipb2MhSRGdw

GRAPH = Y = MSE, X= lambda, horizontal blue line describes the OLS. Dotted line going down describes the variance, dotted line going up describes the bias. Solid U line skewed to the right shows shrinkage.

To find the minimum lambda by balancing the two terms we get the following
lambda = (ro*sigma^2)/Sum(betan^2)
Therefore, if all coefficients are large relative to the variance, we set lambda to be small
On other hand if we have a number of small coefficients, we want to pull them closer to 0 by setting lambda to be large.

Charles Stein 1960 . specification for lambda given that we know sigma^2
Stanley Sclove 1960 build on top of the previous model.
In essence they developed a method to set the coefficients to zero if F is greater than a value criteria of c.  So the coefficient is zero if it fails a F-test with a signficance level of c. If it passes the F-test, then we shrink the coefficients by an amount determined by how strongly the F-statitic protest the null hypothesis.

### Multi-level Models

  * **what is an effect:** an effect is anything that influences the value of a response variable at a particular setting of the predictor variables. Effects are translated into model parameters. In linear models, effects become coefficients, representing the proportional contributions of model terms. In nonlinear models, effects often have specific physical interpretations, and appear in more general nonlinear combinations.
  * **Fixed effects:** They represent population parameters, assumed to be the same each time data is collected. Estimating fixed effects is the traditional domain of regression modeling. 
     * Example here....
  * **Random effects:** They are sample-dependent random variables. In modeling, random effects act like additional error terms, and their distributions and covariances must be specified.
     * Example here....
  * **Mixed-effects models** account for both fixed and random effects. As with all regression models, their purpose is to describe a response variable as a function of the predictor variables. Mixed-effects models, however, recognize correlations within sample subgroups. In this way, they provide a compromise between ignoring data groups entirely and fitting each group with a separate model.
