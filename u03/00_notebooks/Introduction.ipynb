# Introduction
All neuroscience is a problem of source separation. 

We have many phenomenas happening in the brain simultaneously and they generate many observable signals. Our challenge is:
1. to detect these signals
2. to separate the signal from a specific source of interrest from the other sources (noise).

There are multiple sources of singals and multiple sources of noise and because our equipment has limited capacity, all these data is recorded together.

The way we observe these signals is through measuring mechanisms, usually some sensor, test, task or questionaire. This measuring sistem acquire the manifest variables from multiple sources. They can be true sources (like a neuron, or a lever press) or a latent constructs (intelligence, attention or conscience). 

NOTE: sources have different degrees of objectivity. The closer the source it is from the manifest variable, the more objective it is, the farther it is  the more it is a latent construct.

The challenge is to combine multiple measurements of manifest variables to try to separate the source and do some estimation about what is happening in the true source.

True sources -> to measurement -> sources (recovered)

How sources can be separated?
- Anatomical source separation: Just study one area of the brain
- Cognitive source separation: Focus on one cognitive aspect.
- Temporal source separation
- Spatial source separation
- Statistical source separation

Temporal and spatial filtering
Filtering means to weight your data (windowise temporal or spatial) by some function (kernel, filter or template) that transforms the data so that some characteristics are highlighted and other suppressed. 

So every point in the output of the filtering process is kernel weighted sum of the original data.

Visual example of running average, visual example of smoothing, edge detection.

You can have a combination of spatio-temporal filters. Where you have two kernels/filters interacting one across time within each sensor and other across space between multiple sensors. But the concept is exactly the same.

## Spectral separation
Power spectrum (x= frequencies, y= power in arbitrary units)
It is possible to separate sources in the spectral space if the sources do not overlab in their spectral representation. But if they do overlap, it is possible to remove most of the noise but not completely remove the source of noise. Thus, the assumption to use spectral methods is that different sources have different spectral signatures.

### how eeg works
Neurons generate electrical field that can be detected with sensors. In EEG the electrode sits on the top of the skull.
The collective electric field of many neurons produce the signal that can be detected by the eeg sensor.

this gives us a sense that EEG reflects the activity of the electric cells in the brain (neurons). **Explain the mechanism of LFP** But it is not clear about what they mean.

Curiously, the relatively slow and collective signals captured by EEG are not the kind of activity that is employed to generate cognition and behavior, that comes from much finer grained, faster and spatio-temporally constrained activity of action potentials.

##### advantages
- Electrophysiology is good because you can measure directly the medium of communication between the cells in the brain.
- It has high temporal resolution, so it can capture the activity of the brain as cognition occurs
- Richness of data allows for physiologically inspired analysis( Oscillations, synchronizations, connectivity, complexity/scale-free)
- It allows to link results across scales/methods/species (cohen et al, 2014) Because electrical activity can be measured across multiple scales.
Buzsaki et al., 2013 Neuron (plot with frequency peaks across different species)

##### disadvantages
- it is limited to synchronous large scale potentials
- geometrically opposed electrical fields will cancel eachother. Thus we wont be able to detect the activity, even if the activity is strong. So there are many things that are happening in the brain that is invisible to eeg.
- there is uncertainty regarding the anatomical localization source. Dalal et al 2008 NeuroImage
- The data is very complex, time-consuming and annoying.
- High temporal precision is a disadvantage when you are studying slow cognitive process and you dont know exaclty when it happens

## ICA independent component analysis for data cleaning
Decomposes the multi-channel signal into a series of components that are maximally independent from eachother. 

1. perform ICA
2. look at the first ICAs that seem to capture variance
3. Inspect the dynamic to tell the difference between artificial and real signals. Look for task related activity.
4. Inspect topology to tell if it originates from the brain or other part of the body. Look for smoothness, distribution and center of the brain.
5. Decide if it is beter to keep ro remove the component.
Components that are distributed smoothly across multiple electrodes are a good indication of a real signal. While concentrated in one, are indication of artifact.
Remove as little as possible. The eyeblink is one that must be removed.

rebuild the data using the selected components. and compare how the data look now when compared to the raw data. It should look the same in epochs without the artifact and look different in epochs with the artifact.

Remember that the only way to have clean data is by simulating data. Real data is always dirty. Try to clean as little as possible. And try to have a lot of data so you are not very concerned about one missing/bad data point.

## Topographical maps
Topographical maps show how activity is distributed over the scalp. It describes the Spatial distribution of electrodes on the scalp and their associated activity.

Each black dot is one electrode and the activity is only measured at that particular point. 

There is the assumption that the signal changes smoothly from one electrode to the next one. Consequently, it is possible to estimate the activity at any point on the scalp by interpolating the activity between two or more electrodes.

The topographical map can be associates with time, by taking the map at different points in time.

Flat-topographical maps and head plots (3D) - needs to show to or three angles to be able to show the entire scalp.

## Overview of time-domain analyses (ERPs)
ERPs or event related potential (look for that in books)
ERP - Time-frequency - spectral - connectivity

ERP is the average of the signal across different trials aligned to a specific event.

It loses trial specific dynamic that might have relevant information. 

Also, there is phase-locked and phase-unlocked activities that might influence your average.

The motivation for ERPs can derive from the cognitivist model (Roesler 2015, History of psychology) where between the input and the output there would happen a sequence of operations in the brain that would be observable by a sequential pattern in the dynamic of the activity. So, by manipulating the stimuli and observing the behavior and the EEG (E-Phys signal) one could be able to infere the intervening processes.

**example where ERP average in single trial has information but becomes a flat line after 6 trials** The information is lost because of non-phase-locked dynamics. When the deflections have slight different timing, they produce destructive interference. Now, the average in the power frequency domain still preserve the information. (trial 1 to current)

### Phase-locked vs non-phase-locked activity
Uhlhaas and singer nature reviews. Difference between evoked (phase locked) and induced (non-phase locked) activity.
If you generate the ERP and than take the frequency power spectrum analysis you lose information of non-phase locked activity. Instead, what you should do is to perform the power spectrum analysis of every trial and than take the average.

time-locked vs non-time locked
phase-locked vs non phase locked.
If:
- time locked and phase locked: see in single trials, in trial average, ERP and time frequency power
- if time locked and non phase locked: trial average is reduced, the time frequency power is visible and ERP is lost
- non-time locked and non phase locked: same as above
- phase locked bin time locked : undefined

#### How to inspect time frequency results like a pro
1. **Determine what is being shown in the plot**. Power, connectivity, correlation with behavior, etc.
2. **Inspect the ranges and limits of the plot**. what are the x- and y-axes? color scale? is activity cut off at the edges?
3. **Look at the results**. Distributed or localized? Why only this electrode or time/ frequency window? Is there something that was cut from the plot?
4. **Link the results to the experiment design**. what is the time=0? Multiple task events?
5. **Understand the stantistical procedures**. Thresholding? Multiple comparisons addressed? exploratory?
